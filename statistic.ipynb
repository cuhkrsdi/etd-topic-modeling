```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#df = pd.read_csv("data/statistic.csv")
statistic_url = "https://raw.githubusercontent.com/cuhkrsdi/etd-topic-modeling/main/data/statistic.csv"
df = pd.read_csv(statistic_url)
```

## Proportion of Hong Kong related thesis in ETD collection.

```python
df['proportion'] = df.hk_related / df.total * 100
```

```python
# The proportion of Hong Kong related thesis in ETD collection. Line plot from 1982 to 2021.
# Extract the data from 1982 to 2021, since data size before 1982 is too slow, and is not complete after 2021.
plt.figure(figsize=[10, 5])
ax = sns.lineplot(data=df[14:54], x='year', y='proportion')
ax.set(title='Proportion of Hong Kong Related Thesis in ETD Collection',
       xlabel='Year',
       ylabel='HK-related Thesis (Percentage)')
```

## Compare the trends 

```python
df_melt = df[['year', 'hk_related', 'total']][14:54]
df_melt = df_melt.melt(['year'], var_name='category', value_name='count')
```

```python
df_melt['category'] = df_melt['category'].replace('hk_related', 'HK Related')
df_melt['category'] = df_melt['category'].replace('total', 'Total')
df_melt.columns = ['Year', 'Category', 'Number of Thesis']
```

```python
plt.figure(figsize=[10, 5])
ax_2 = sns.lineplot(data=df_melt, x='Year', y='Number of Thesis', hue='Category')

ax_2.set(title='Compared the Trends of Hong Kong Related and Total Thesis',)
```
